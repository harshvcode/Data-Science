```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df
#This model will predict whether a person will surive on the basis of Pclass, Sex, Age, Fare
```

```python
inpu = df.drop(['Survived', 'PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
inpu
```

```python
targ = df['Survived']
targ
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
inpu['Pclass_n'] = le_Pclass.fit_transform(inpu['Pclass'])
inpu['Sex_n']= le_Pclass.fit_transform(inpu['Sex'])
inpu['Age_n']= le_Pclass.fit_transform(inpu['Age'])
inpu['Fare_n']= le_Pclass.fit_transform(inpu['Fare'])
inpu
```

```python
inpu_n = inpu.drop(['Pclass', 'Sex','Age', 'Fare'], axis = 'columns')
inpu_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inpu_n, targ)
```
